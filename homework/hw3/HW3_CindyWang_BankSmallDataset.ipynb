```python

# Code categorical independent variables of two possible value
bank_s['default_b'] = bank_s.default.map(lambda x: 1 if x == 'yes' else 0)
bank_s['housing_b'] = bank_s.housing.map(lambda x: 1 if x == 'yes' else 0)
bank_s['loan_b'] = bank_s.loan.map(lambda x: 1 if x == 'yes' else 0)

# Code categorical independent variables of more than two possible value
dummies_job = pd.get_dummies(bank_s.job, prefix='job').ix[:, 1:]
dummies_marital = pd.get_dummies(bank_s.marital, prefix='marital').ix[:, 1:]
dummies_education = pd.get_dummies(bank_s.education, prefix='education').ix[:, 1:]
dummies_contact = pd.get_dummies(bank_s.contact, prefix='contact').ix[:, 1:]
dummies_month = pd.get_dummies(bank_s.month, prefix='month').ix[:, 1:]
dummies_poutcome = pd.get_dummies(bank_s.poutcome, prefix='poutcome').ix[:, 1:]


X= pd.concat([bank_int_df, dummies_job,dummies_marital,
                          dummies_education,dummies_contact,dummies_month,
                          dummies_poutcome,
                          bank_s['default_b'],
                          bank_s['housing_b'],
                          bank_s['loan_b']
                          ], axis=1)
y = bank_s.y_binary

print 'final type:', type(X)
print X.head()

print 'Prior dummy shape:', bank_s.shape
print 'Post dummy shape:', X.shape
```

```python
# split into train/test
from sklearn.cross_validation import cross_val_score
from sklearn.tree import DecisionTreeClassifier
import seaborn as sns

x_axis = []
y_axis = []
for k in range(20)[2:]:
    dtc = DecisionTreeClassifier(max_depth=k, random_state=1)
    scores = cross_val_score(dtc, X, y, cv=10, scoring='accuracy')
    #print 'For max depth:', k
    x_axis.append(k)
    #print 'Decision tree accuracy: {}'.format(np.mean(scores))
    y_axis.append(np.mean(scores))

print x_axis
print y_axis

ig, ax = plt.subplots(1,1)
ax.plot(x_axis, y_axis, color='red', label = 'Decision Tree Bank Small dataset')
ax.set_ylabel('Accuracy Rate')
ax.set_xlabel('Max Tree Depth')
ax.legend(loc='upper right',frameon=True)
```

```python
# Predict based on final model
treeclf = DecisionTreeClassifier(max_depth=7, random_state=1)
treeclf.fit(X, y)
feature_cols = X.columns
pd.DataFrame({'feature':feature_cols, 'importance':treeclf.feature_importances_}).sort(['importance'], ascending=False)
```

```python
from sklearn.tree import export_graphviz

# visualize the tree
with open("Images\hw3_banksmall.dot", 'wb') as f:
    f = export_graphviz(treeclf, out_file=f, feature_names=feature_cols)
```
